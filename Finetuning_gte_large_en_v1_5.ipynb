```python
from collections import defaultdict

for fn_name, fn_list in all_function_lists.items():
    print(f"Checking {fn_name} ({len(fn_list)} samples)")

    descriptions = set()

    for sample in fn_list:
        assert len(sample) == 3, f"{fn_name}: bad sample length"

        prompt, func_desc, score = sample

        assert isinstance(prompt, str)
        assert isinstance(func_desc, str)
        assert isinstance(score, float)
        assert 0.0 <= score <= 1.0, f"{fn_name}: invalid score {score}"

        descriptions.add(func_desc)

    assert len(descriptions) == 1, (
        f"{fn_name} has multiple descriptions!"
    )
```

```python
from sentence_transformers import InputExample

examples = [
    InputExample(
        texts=[prompt, func_desc],
        label=float(score)
    )
    for prompt, func_desc, score in all_data
]
```

```python
from sklearn.model_selection import train_test_split

train_examples, val_examples = train_test_split(
    examples,
    test_size=0.1,      # 90% train / 10% val
    random_state=42,
    shuffle=True
)
```

```python
len(train_examples), len(val_examples)
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(
    train_examples,
    shuffle=True,
    batch_size=32   # 32 if GPU allows
)

val_dataloader = DataLoader(
    val_examples,
    shuffle=False,
    batch_size=32
)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "Alibaba-NLP/gte-large-en-v1.5",
    device="cuda",                  # or "cpu"
    trust_remote_code=True           # <-- REQUIRED
)
```

```python
print(model.device)
print(model.get_sentence_embedding_dimension())
```

```python
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator

val_sentences1 = [ex.texts[0] for ex in val_examples]
val_sentences2 = [ex.texts[1] for ex in val_examples]
val_scores     = [ex.label    for ex in val_examples]

evaluator = EmbeddingSimilarityEvaluator(
    val_sentences1,
    val_sentences2,
    val_scores,
    name="shell-val"
)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# 1️⃣ Load fine-tuned model
model = SentenceTransformer(
    "./gte_shell_finetuned",
    trust_remote_code=True
)

# 2️⃣ Canonical function descriptions (use EXACT ones from training)
functions = {
    "shell_open": "shell_open: Opens applications, files, or system resources.",
    "shell_clean": "shell_clean: Cleans temporary files, cache, and unused system data.",
    "shell_make": "shell_make: Creates files, folders, or new system resources.",
    "shell_read": "shell_read: Reads files or displays file contents.",
    "shell_search": "shell_search: Searches files, folders, or information.",
    "shell_news": "shell_news: Fetches latest news and headlines.",
    "shell_weather": "shell_weather: Provides weather information and forecasts.",
    "shell_stocks": "shell_stocks: Fetches stock prices and market data.",
    "shell_download": "shell_download: Downloads files from the internet."
}

fn_names = list(functions.keys())
fn_descs = list(functions.values())

# 3️⃣ Pre-embed function descriptions
fn_embeddings = model.encode(
    fn_descs,
    normalize_embeddings=True
)

# 4️⃣ Test examples (ALL intended for the SAME function)
test_queries = [
    "open chrome",
    "clean the junk in temp folder",
    "start firefox",
    "open system settings",
    "downlaod gta 5 from https://dodirepacks.com"
]

print("=== Cosine similarity predictions ===\n")

for query in test_queries:
    q_emb = model.encode(query, normalize_embeddings=True)

    scores = cosine_similarity(
        [q_emb],
        fn_embeddings
    )[0]

    best_idx = scores.argmax()

    print(f"Query: {query}")
    print(f"  Predicted function: {fn_names[best_idx]}")
    print(f"  Cosine score: {scores[best_idx]:.3f}\n")
```
